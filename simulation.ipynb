## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
```

## Creating Game of Life simulation

```python
t = 100
size = 50

random = np.random.choice([0, 1], size=(size, size)) 

def game_of_life(t, size, grid):
    plot_hist = [grid]
    for i in range(t):
        next_state = np.zeros((size, size)) 
        for x, y in np.ndindex(grid.shape):
            # Complete the function to implement the rules of Game of Life to set the cell values
            plot_hist.append(next_state)
    return np.array(plot_hist)
```

## Visualizaing simulation results

```python
fig = plt.figure() 

sim = game_of_life(t, size, random)

img = plt.imshow(sim[0], cmap="Greys")

def animate(k):
    img.set_array(sim[k])

anim = animation.FuncAnimation(fig, animate, frames=range(t), blit=False)
plt.close()
HTML(anim.to_jshtml())
```
